### **3.4 Spending Trends Over Time**

```python
# Load and preprocess data
data = pd.read_csv('April_2024-October_2024.csv')
data['Date'] = pd.to_datetime(data['Date'], format='%d/%m/%Y', errors='coerce')
# Transaction Timing
spending_by_month = data.groupby(data['Date'].dt.to_period('M'))['Amount'].sum()
plt.figure(figsize=(12, 6))
spending_by_month.plot(color='green', marker='o')
plt.title('Monthly Spending Trends')
plt.xlabel('Month')
plt.ylabel('Total Amount')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

1. **Fluctuations in Spending:**  
   - Spending demonstrates significant variation across months, emphasizing dynamic financial activity over the analyzed period.
